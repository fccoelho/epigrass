```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
import panel as pn
import panel.widgets as pnw
import hvplot.pandas
import holoviews as hv
import geoviews as gv
import param
import glob, os
```

```python
pn.extension()
```

```python
material = pn.template.MaterialTemplate(title='Epigrass Dashboard')

pn.config.sizing_mode = 'stretch_width'

xs = np.linspace(0, np.pi)
freq = pn.widgets.FloatSlider(name="Frequency", start=0, end=10, value=2)
phase = pn.widgets.FloatSlider(name="Phase", start=0, end=np.pi)


@pn.depends(freq=freq, phase=phase)
def sine(freq, phase):
    return hv.Curve((xs, np.sin(xs * freq + phase))).opts(
        responsive=True, min_height=400)


@pn.depends(freq=freq, phase=phase)
def cosine(freq, phase):
    return hv.Curve((xs, np.cos(xs * freq + phase))).opts(
        responsive=True, min_height=400)


class SimMap(param.Parameterized):
    pth = '../../demos/outdata-rio'
    maps = glob.glob(os.path.join(pth, '*.gpkg'))
    selector = param.ObjectSelector(maps)

    # label = param.(pth)
    # simulation =

    @param.depends('pth', watch=True)
    def _dir_update(self):
        maps = glob.glob(os.path.join(self.pth, '*.gpkg'))
        self.param['selector'].objects = maps
        self.selector = maps[0]

    @param.depends('selector')
    def view(self):
        print(self.selector, type(self.selector), self.maps, os.getcwd())
        if self.selector:
            df = gpd.read_file(self.maps[0])
            f = gv.Polygons(df, vdims=['prevalence', 'totalcases', 'arrivals'], label='Prevalence').opts(
                width=600,
                tools=['hover'],
                toolbar='above',
                colorbar=True
            )
            return f
        else:
            return hv.DynamicMap(sine)

sm = SimMap()
material.sidebar.append(freq)
material.sidebar.append(phase)
material.sidebar.append(sm)
```

```python
material.main.append(
    pn.Column(
        pn.Row(
            pn.Card(sm.view)
        ),
        pn.Row(
            pn.Card(hv.DynamicMap(sine), title='Sine'),
            pn.Card(hv.DynamicMap(cosine), title='Cosine')
        )
    )
)
```
